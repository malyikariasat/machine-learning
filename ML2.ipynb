**Dealing with data (Inconsistencies)**

```python
# import required libraries
import pandas as pd
```

```python
data = {
    'date':['2025-12-01','01-12-2025','2025/12/01','12-01-2025'],
    'country':['USA','U.S.A','America','United States'],
    'name':['John Doe','Jonh Doe','Jane Doe','Jane Doe'],
#    'name':['Iqbal','Iqbaal','Munawar','Munawwar'],
    'sales_2024':[100,200,None,200],
    'sales_2025':[None,150,300,150],
        }
df = pd.DataFrame(data)
df.head()
```

```python
# standardizing the date format
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d', errors='coerce')
df
```

```python
# correct the name of the country
country_mapping = {
    'USA': 'Unitd Stats of America',
    'U.S.A': 'Unitd Stats of America',
    'America': 'Unitd Stats of America',
    'United States': 'Unitd Stats of America'
}
df['country'] = df['country'].replace(country_mapping)
df
```

```python
# correct typographical mistakes in name
# lets assume we want to correct "Jonh Doe" with "John Doe"
df['name'] = df['name'].str.replace('Jonh Doe', 'John Doe')
#df['name'] = df['name'].replace({'Iqbaal':'Iqbal','Munawar':'Munawwar'})

df
```

```python
# remove duplicates
df = df.drop_duplicates()
df
```

```python
# remove duplicates
df = df.drop_duplicates(subset='country')
df
```

```python
# resolving contradictory data
# sales of 2025 should be greater then 2024
df = df.drop(df[df['sales_2024']<=df['sales_2025']].index)
df
```

```python
# Simple Linear Regression Model in Python

from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Example data: X = input, y = output
X = np.array([[1], [2], [3], [4], [5]])  # Features (2D)
y = np.array([2, 4, 6, 8, 10])          # Target (1D)
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Make a prediction
prediction = model.predict([[9]])

print("Predicted value for input 9:", prediction[0])
```
