## **IMPORT LIBRARIES**

```python
# import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load data

```python
data = sns.load_dataset('titanic')
data.head()
```

Visualiza data

```python
# visualize the data
plt.figure(figsize=(10, 8))
sns.heatmap(data.isnull(), cbar=False)
plt.show()
```

```python
# calculate missing values
print("----------------------------------")
print("Missing values in each column")
print("----------------------------------")
print(data.isnull().sum())
```

```python
# sort values
data.isnull().sum().sort_values(ascending=False)
```

```python
# get data information
data.info()
```

```python
# get percentage of null values in each column
(data.isnull().sum()/len(data)*100).sort_values(ascending=False)
```

```python
# get percentage of null values in each column uptill two decimal digit
round((data.isnull().sum()/len(data)*100),2).sort_values(ascending=False)
```

```python
# mean of age to fill missing values in age column
round(data['age'].mean(),2)
```

```python
# mean of age to fill missing values in age column
data['age'] = data['age'].fillna(data['age'].mean())
```

```python
# drop deck column
data.drop('deck', axis=1, inplace=True)
```

```python
# replace embarked missing values with mode (Home Work)
data['embark_town'].value_counts()
data['embark_town'].mode()
data['embark_town'] = data['embark_town'].fillna(data['embark_town'].mode()[0])
data['embarked'] = data['embarked'].fillna(data['embarked'].mode()[0])
```

```python
# check null values
round((data.isnull().sum()/len(data)*100),2).sort_values(ascending=False)
```
